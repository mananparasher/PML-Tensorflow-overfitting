```
import pathlib
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
import os
from tensorflow.keras import regularizers
```

```
path = keras.utils.get_file("poker-hand-training-true.data", "https://archive.ics.uci.edu/ml/machine-learning-databases/poker/poker-hand-training-true.data")
path
```

```
column_names = ['Suit_1','Rank_1','Suit_2','Rank_2','Suit_3','Rank_3','Suit_4','Rank_4','Suit_5','Rank_5','Class']
dataset = pd.read_csv(path, names=column_names,
                      na_values = "?", comment='\t',
                      sep=",", skipinitialspace=True)

dataset.head(5)
```

```
y=dataset.pop("Class")
X=dataset.copy()
```

```
model = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation='relu', input_shape=[len(X.keys())]),
    tf.keras.layers.Dense(10,activation="softmax")
  ])
```

```
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                optimizer='adam',
                metrics=['accuracy'])
```

```
model.fit(X,y,epochs=100,validation_split=0.5)
```

```
Regulazrization
```

```
regularized_model = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation='relu',kernel_regularizer=regularizers.l2(0.001), input_shape=[len(X.keys())]),
    tf.keras.layers.Dense(50, activation='relu',kernel_regularizer=regularizers.l2(0.001), input_shape=[len(X.keys())]),
    tf.keras.layers.Dense(10,activation="softmax")
  ])
```

```
regularized_model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                optimizer='adam',
                metrics=['accuracy'])
```

```
regularized_model.fit(X,y,epochs=100,validation_split=0.5)
```

```
Dropurates
```

```
droputrate_model = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation='relu', input_shape=[len(X.keys())]),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(50, activation='relu', input_shape=[len(X.keys())]),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10,activation="softmax")
  ])
```

```
droputrate_model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                optimizer='adam',
                metrics=['accuracy'])
```

```
droputrate_model.fit(X,y,epochs=100,validation_split=0.5)
```
